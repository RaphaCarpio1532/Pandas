```python
import pandas as pd
```

### Carica il DataFrame

```python
df_books = pd.read_csv('/content/bestsellers-with-categories_.csv',sep=',',header=0)
df_books.head(2)
```

### Utilizzo di GroupBy per Analisi di Base:

 groupby()

```python
df_books.groupby('Author').count().head(4)
```

```python
df_books.groupby('Author').mean(numeric_only=True).head(4)
```

```python
df_books.groupby('Author').max().head(4)
```

```python
df_books.groupby('Author').sum(numeric_only=True).head(4)
```

```python
df_books.groupby('Author').sum(numeric_only=True).loc['Abraham Verghese']
```

```python
df_books.groupby('Author').sum().reset_index()
```

# Applicazione di Funzioni di Aggregazione Specifiche:



```python
df_books.groupby('Author').agg(['min','max'])
```

```python
df_books.head(2)
```

```python
df_books.groupby('Author').agg({'Reviews':['min','max'],'User Rating':'sum'})
```
