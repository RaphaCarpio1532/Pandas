# **Pivot & Melt**

# **Pivot**

```python
import pandas as pd
```

**Carica il DataFrame che abbiamo usato nel corso:**

```python
df_books = pd.read_csv('/content/bestsellers-with-categories_.csv', sep=',', header=0)
```

Esploralo vedendo le sue prime 5 righe:

```python
df_books.head(5)
```

**Applica pivot_table:**



***1. Pivot Table per Categoria e Anno***


```python
df_books.pivot_table(index='Genre', columns='Year', values='User Rating', aggfunc='mean')
```

 ***2. Pivot Table con Multi-Index
Analizziamo i prezzi medi dei libri per autore e genere.***

```python
df_books.pivot_table(index=['Author', 'Genre'], values='Price', aggfunc='mean')
```

***3. Pivot Table con Valori Multipli***
 * Esaminiamo sia la media sia il conteggio dei rating degli utenti.



```python
df_books.pivot_table(index='Genre', values='User Rating', aggfunc=['mean', 'count'])
```

***4. Pivot Table con Marginali***
  * Aggiungiamo i totali di riga e colonna per avere una visione complessiva.



```python
df_books.pivot_table(index='Genre', columns='Year', values='User Rating', aggfunc='mean', margins=True)
```

# **Melt**

```python
df_books[['Name','Genre']].head(5)
```

***1. Melt di Base***
* Convertiamo le colonne 'Author' e 'Genre' in righe.

```python
df_books[['Author', 'Genre']].head().melt()
```

***2. Melt con Colonne Multiple***
  
  * Fondiamo piÃ¹ colonne, mantenendo 'Author' come identificatore.

```python
df_books.melt(id_vars='Author', value_vars=['Year', 'User Rating'])
```

**3. Melt con Nomi di Colonne Personalizzati**

* Personalizziamo i nomi delle colonne risultanti.




```python
df_books.melt(id_vars='Author', value_vars='Genre', var_name='Categoria', value_name='Valore')
```

***4. Melt su Tutto il DataFrame***

  * Applichiamo melt all'intero DataFrame per trasformare ogni colonna in una riga.



```python
df_books.melt()
```

```python
melted_df = df_books.melt()
for name, group in melted_df.groupby('variable'):
    print(f"--- Prime 5 righe per la colonna originale '{name}' ---")
    print(group.head())
```
