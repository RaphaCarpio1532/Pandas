### Importar librer√≠as

```python
import pandas as pd
```

### Carica DataFrame

```python
df_books = pd.read_csv('/content/bestsellers-with-categories_.csv')
df_books.head(2)
```

### Definire la funzione

```python
def two_times(value):
    return value * 2
```

### Apply()

```python
df_books['Rating_2'] = df_books['User Rating'].apply(two_times)
```

```python
df_books.head()
```

***Le funzioni lambda***

```python
double = lambda x: x * 2
print(double(5))  # Output: 10
```

```python
numbers = [1, 2, 3, 4]
doubled_numbers = map(lambda x: x * 2, numbers)
print(list(doubled_numbers))  # Output: [2, 4, 6, 8]
```

```python
df_books['Rating_2'] = df_books['User Rating'].apply(lambda x : x * 3)
```

```python
df_books.head(15)
```

***Applicazione Condizionale:***

```python
df_books['Rating_2'] = df_books.apply(lambda x : x['User Rating'] * 2 if x['Genre'] == 'Fiction' else x['User Rating'],axis=1)
```
