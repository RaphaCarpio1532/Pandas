
### Importar Librerie

```python
import pandas as pd
import numpy as np
```

### Crea dizionario e DataFrame

```python
dict = {'Col1':[1,2,3,np.nan],
 'Col2':[4,np.nan,6,7],
 'Col3':['a','b','c',None]}
```

```python
df = pd.DataFrame(dict)
```

```python
df
```

### Ver valores nulos

```python
#Restituisce True Ogni volta che viene trovato un valore nullo all'interno del Dataframe
df.isnull()
```

```python
#Possiamo portare questo a un metodo numerico
#Questo è utilizzato in molti aspetti, tensori nelle reti neurali

df.isnull()*1
```

### Riempi valori nulli

```python
#Viene utilizzato per riempire i valori nulli con "Missing
df.fillna('Missing')
```

```python
#Viene utilizzato per riempire i valori nulli con la media
#L'avviso ci spiega che non è possibile calcolare la media delle lettere "a,b,c"

df.fillna(df.mean())
```

### Interpolare

```python
#qui stai assumendo con che valori dovrebbe continuare di lavorare con una serie.
df.interpolate()

#È un modo per gestire i valori nulli purché sappiamo che i dati sono
#strutturati con una serie e una frequenza esatta, come questo tipo di dati.
df
```

```python
df.interpolate()
```

### Borrar valores nulos

```python
df.dropna()
```

```python
#Per eliminare valori nulli da una colonna specifica
df.dropna(subset=["Col2"])
```
