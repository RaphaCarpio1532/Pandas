
# join() - Corrispondenza degli Indici

```python
# Creazione del DataFrame Squadra Azzurra
data_azzurra = {
    'Nome': ['Marco', 'Alessandro', 'Matteo'],
    'Posizione': ['Portiere', 'Difensore', 'Centrocampista'],
    'Squadra': ['Juventus', 'Milan', 'Roma']
}
df_azzurra = pd.DataFrame(data_azzurra, index=['k0', 'k1', 'k2'])

# Creazione del DataFrame Squadra Bianca
data_bianca = {
    'Nome': ['Giovanni', 'Francesco', 'Gabriele'],
    'Posizione': ['Portiere', 'Centrocampista', 'Attaccante'],
    'Squadra': ['Napoli', 'Fiorentina', 'Atalanta']
}
df_bianca = pd.DataFrame(data_bianca, index=['k0', 'k2', 'k3'])

# Eseguire il join (How, Left,right, Outer, inner)

```

```python
df_azzurra.join(df_bianca,how='inner')
```

```python
df_azzurra.join(df_bianca, how='left', lsuffix='_Azzurra', rsuffix='_Bianca')
```

# Left Join


```python
left_join = pd.merge(df_azzurra, df_bianca, on='Posizione', how='left', suffixes=('_Azzurra', '_Bianca'))
```

```python
left_join
```

# Right Join


```python
right_join = pd.merge(df_azzurra, df_bianca, on='Posizione', how='right', suffixes=('_Azzurra', '_Bianca'))
```

```python
right_join
```

# Inner Join


```python
inner_join = pd.merge(df_azzurra, df_bianca, on='Posizione', how='inner', suffixes=('_Azzurra', '_Bianca'))
inner_join
```

# Outer Join


```python
outer_join = pd.merge(df_azzurra, df_bianca, on='Posizione', how='outer', suffixes=('_Azzurra', '_Bianca'))
outer_join
```
